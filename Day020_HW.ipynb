```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(r.text, 'html.parser')
table = soup.find('table', class_='TABLE_G')
print(table)
```

```python
d={}
for i in table.find_all('td'):
    if 'style' in i.attrs:
        a=i.text
        d[a]={}
    elif 'valign' not in i.attrs:
        if i.text.strip()!='':
            if '/' in i.text:
                date=i.text
            else:
                d[a][date]=i.text
d
```
