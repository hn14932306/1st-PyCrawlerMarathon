### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup
import datetime
import pandas as pd
columns=["日期","類別","標題"]
today = datetime.date.today().strftime("%Y/%m/%d")
url='https://www.ettoday.net/news/news-list.htm'
r=requests.get(url)
soup=BeautifulSoup(r.text,"html5lib")
articles=[]
for i in soup.find(class_="part_list_2").find_all('h3'):
    if((i.find(class_="date").text)[:10]==today):
        articles.append([i.find(class_="date").text,i.find('em').text,i.find_all('a')[-1].text])
        print(i.find(class_="date").text,i.find('a').text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
df=pd.DataFrame(articles,columns=columns)
df
```

### ③ 哪一個類別的文章最多

```python
df1=pd.DataFrame(df.groupby('類別').size(),columns=['文章數量'])
df1['排名']=df1.rank(method='min', ascending=False)
df1.sort_values(by=['排名'])
```
