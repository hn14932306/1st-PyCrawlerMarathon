### 根據範例檔案的結果：

1. 請問高雄市有多少地區有溫度資料？
2. 請取出每一個地區所記錄的第一個時間點跟溫度
3. 請取出第一個地區所記錄的每一個時間點跟溫度

```python
# 1. 請問高雄市有多少地區有溫度資料？

import xmltodict
fh = open("./data/64_72hr_CH.xml", "r")
xml = fh.read()
fh.close()
d = dict(xmltodict.parse(xml))
location = d['cwbopendata']['dataset']['locations']['location']
print(len(location))
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度
for i in location:
    print(i['locationName'])
    print(i['weatherElement'][0]['time'][0]['dataTime'])
    print(i['weatherElement'][0]['time'][0]['elementValue']['measures'],i['weatherElement'][0]['time'][0]['elementValue']['value'])
```

```python
# 3. 請取出第一個地區所記錄的每一個時間點跟溫度
location1=d['cwbopendata']['dataset']['locations']['location'][0]['weatherElement'][0]['time']
for i in range(0,len(location1)):
    print(d['cwbopendata']['dataset']['locations']['location'][0]['weatherElement'][0]['time'][i]['dataTime'])
    print(d['cwbopendata']['dataset']['locations']['location'][0]['weatherElement'][0]['time'][i]['elementValue']['measures'],d['cwbopendata']['dataset']['locations']['location'][0]['weatherElement'][0]['time'][i]['elementValue']['value'])
```
