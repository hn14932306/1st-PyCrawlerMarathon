## 1. 取出今天所有的新聞

```python
import requests
from bs4 import BeautifulSoup

from selenium import webdriver
browser = webdriver.Chrome(executable_path='chromedriver')
browser.get("https://www.ettoday.net/news/news-list.htm")
from selenium.webdriver.support.ui import Select
select_year= Select(browser.find_element_by_id('selY'))
select_year.select_by_value('2020')

select_month= Select(browser.find_element_by_id('selM'))
select_month.select_by_value('1')
select_day= Select(browser.find_element_by_id('selD'))
select_day.select_by_value('18')
browser.find_element_by_id('button').click()
import time
for i in range(50):
    browser.execute_script("window.scrollTo(0, 100000)")
    time.sleep(1)
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")
from selenium import webdriver
from bs4 import BeautifulSoup
```

## 2. 取出現在時間兩小時內的新聞

```python
import re
from datetime import datetime
from datetime import timedelta
today=datetime.now()
fmt = '%Y/%m/%d %H:%M'
time=re.split(' |-',str(datetime.now()))[3][:5]
for i in soup.find('div',class_='part_list_2').find_all('h3'):
    if '2020/01/18'==i.find(class_='date').text[:10]:
        if (datetime.now() - datetime.strptime(i.find(class_="date").text, fmt)) <= timedelta(hours=2):
            print(i.find(class_="date").text, i.find_all('a')[-1].text)
            
```

## 3. 根據範例，取出三天前下午三點到五點的新聞

```python
import time
day = datetime.now().day-3
select_day= Select(browser.find_element_by_id('selD'))
select_day.select_by_value(str(day))
browser.find_element_by_id('button').click()
for i in range(20):
    browser.execute_script("window.scrollTo(0, 100000)")
    time.sleep(2)
html_source = browser.page_source
soup = BeautifulSoup(html_source, "html5lib")
for i in soup.find(class_="part_list_2").find_all('h3'):
    news_datetime = datetime.strptime(i.find(class_="date").text, fmt)
    if news_datetime.date().day == day:
        if 15 <= news_datetime.hour <= 17:
            print(i.find(class_="date").text, i.find_all('a')[-1].text)
```
